```python
import librosa
import numpy as np
import soundfile as sf
import parselmouth
import torch
from bark import generate_audio as bark_generate_audio
from bark.generation import preload_models
from audiocraft.models import MusicGen
import difflib
import whisper
import os

TARGET_LOUDNESS = {
    "happy": {"vocals": -12.0, "music": -18.0},
    "sad": {"vocals": -18.0, "music": -20.0},
    "angry": {"vocals": -2.0, "music": -14.0},
    "romantic": {"vocals": -15.0, "music": -18.0},
    "excited": {"vocals": -10.0, "music": -13.0},
    "haunting": {"vocals": -17.0, "music": -19.0}
}

def generate_music_with_melody(mood, lyrics, duration=12, speed_factor=1.0):
    print(f"🎼 Generating {mood} background music with melody for the song...")
    model = MusicGen.get_pretrained('facebook/musicgen-small')
    model.set_generation_params(duration=duration)

    mood_descriptions = {
        "happy": "funky groove with brass instruments and energetic bassline",
        "sad": "slow emotional piano melody",
        "angry": "angry tone",
        "romantic": "soft acoustic melody with gentle rhythm and emotional tone",
        "excited": "fast-paced electronic melody with bright synths",
        "haunting": "dark ambient melody with eerie tones and slow pads"
    }
    description = mood_descriptions.get(mood.lower(), "ambient instrumental")
    wav = model.generate([description])[0].cpu().numpy()

    if not np.all(np.isfinite(wav)):
        print("⚠️ Detected NaNs or infs in generated music. Replacing with zeros.")
        wav = np.nan_to_num(wav)

    if np.max(np.abs(wav)) > 0:
        wav = wav / np.max(np.abs(wav)) * 0.95

    if len(wav.shape) > 1:
        wav = wav[0]

    if speed_factor != 1.0:
        wav = librosa.effects.time_stretch(wav, rate=speed_factor)

    bgm_file = f"{mood}_background_music_with_melody.wav"
    sf.write(bgm_file, wav.astype(np.float32), 32000, format='WAV', subtype='PCM_32')
    print(f"🎧 Background music with melody saved as '{bgm_file}'")
    return bgm_file

def generate_audio_with_bark(lyrics, speed_factor=1.0, pitch_factor=1.0):
    print("🎤 Generating audio using Bark...")
    audio_array = bark_generate_audio(lyrics)

    if speed_factor != 1.0:
        audio_array = librosa.effects.time_stretch(audio_array, rate=speed_factor)
    if pitch_factor != 0.0:
        audio_array = librosa.effects.pitch_shift(audio_array, sr=24000, n_steps=pitch_factor)

    return audio_array

def generate_vocals_with_melody(lyrics, mood="happy", max_attempts=2, speed_factor=1.0, pitch_factor=1.0):
    preload_models()
    best_score = -1
    best_file = None

    for attempt in range(1, max_attempts + 1):
        print(f"\n🎶 Attempt {attempt} for mood '{mood}'...")
        try:
            audio_array = generate_audio_with_bark(lyrics, speed_factor=speed_factor, pitch_factor=pitch_factor)
        except Exception as e:
            print(f"❌ Bark error: {e}")
            continue

        vocals_file = f"{mood}_vocals_attempt{attempt}.wav"
        sf.write(vocals_file, audio_array, 24000)

        autotuned_file = f"{mood}_vocals_attempt{attempt}_autotuned.wav"
        apply_auto_tune(vocals_file, autotuned_file, mood=mood)

        y, sr = librosa.load(autotuned_file, sr=24000)
        rms = check_noise_level(y)
        silence = check_drag(y, sr)
        dragged = detect_dragged_vowels(y, sr)
        transcription = transcribe_audio(autotuned_file)
        similarity = check_similarity(lyrics, transcription)

        score = (similarity * 2) + (1 - min(silence / 2.0, 1)) + (1 - min(rms / 0.2, 1)) + (1 - min(dragged / 2.0, 1))
        print(f"📊 Score: {score:.2f}")

        if score > best_score:
            best_score = score
            best_file = autotuned_file

        if similarity >= 0.75 and silence < 0.6 and rms < 0.1 and dragged < 1.0:
            print("✅ All checks passed!")
            return autotuned_file

    print("\n⚠️ Returning best available attempt.")
    return best_file

def apply_auto_tune(input_file, output_file, mood="happy"):
    snd = parselmouth.Sound(input_file)
    pitch = snd.to_pitch()
    pitch_values = pitch.selected_array["frequency"]
    mean_pitch = np.nanmean(pitch_values[pitch_values > 0])

    target_pitch = {
        "angry": 300,
        "happy": 200,
        "excited": 220,
        "romantic": 190,
        "haunting": 160,
        "sad": 170
    }.get(mood, 180)

    if mean_pitch > 0:
        pitch_shift = target_pitch / mean_pitch
        snd = snd.resample(snd.sampling_frequency * pitch_shift)

    samples = snd.values[0]
    if np.max(np.abs(samples)) > 0:
        samples = samples / np.max(np.abs(samples)) * 0.9
        snd.values[0] = samples

    snd.save(output_file, "WAV")
    print(f"🎵 Auto-Tuned vocals saved as '{output_file}'")
    return output_file

def check_noise_level(audio_array):
    audio_array = np.array(audio_array)
    audio_array = audio_array / np.max(np.abs(audio_array))
    rms = np.sqrt(np.mean(audio_array**2))
    print(f"🔍 RMS (noise level): {rms:.4f}")
    return rms

def check_drag(audio_array, sr, silence_threshold=0.005, frame_duration=0.01):
    frame_length = int(frame_duration * sr)
    hop_length = frame_length
    envelope = np.array([
        np.max(np.abs(audio_array[i:i+frame_length]))
        for i in range(0, len(audio_array), hop_length)
    ])
    silent_frames = envelope < silence_threshold
    max_silence = 0.0
    current = 0.0
    for is_silent in silent_frames:
        if is_silent:
            current += frame_duration
            max_silence = max(max_silence, current)
        else:
            current = 0.0
    print(f"🧘 Longest silence: {max_silence:.2f} sec")
    return max_silence

def detect_dragged_vowels(y, sr, max_syllable_duration=0.6):
    S, _ = librosa.magphase(librosa.stft(y))
    rms = librosa.feature.rms(S=S)[0]
    times = librosa.times_like(rms, sr=sr)

    sustained = rms > np.percentile(rms, 75)
    sustained_regions = []
    start = None

    for i in range(len(sustained)):
        if sustained[i] and start is None:
            start = times[i]
        elif not sustained[i] and start is not None:
            duration = times[i] - start
            if duration > max_syllable_duration:
                sustained_regions.append(duration)
            start = None

    dragged_duration = sum(sustained_regions)
    print(f"🎤 Detected dragged vowels total duration: {dragged_duration:.2f}s")
    return dragged_duration

def check_similarity(original_text, transcribed_text):
    ratio = difflib.SequenceMatcher(None, original_text.lower(), transcribed_text.lower()).ratio()
    print(f"📝 Similarity ratio: {ratio:.2%}")
    return ratio

def transcribe_audio(file):
    model = whisper.load_model("medium")
    result = model.transcribe(file, language="en")
    print(f"📄 Transcription: {result['text']}")
    return result["text"]

def measure_loudness(audio, ref=1.0):
    rms = np.sqrt(np.mean(audio ** 2))
    db = 20 * np.log10(rms / ref + 1e-9)
    return db

def apply_gain(audio, gain_db):
    factor = 10 ** (gain_db / 20)
    return audio * factor

def mix_vocals_and_music(vocal_path, music_path, output_path="final_song_with_melody.wav", mood="happy"):
    print(f"🎚️ Mixing vocals and background music with melody...")
    vocal, sr_vocal = librosa.load(vocal_path, sr=None)
    music, sr_music = librosa.load(music_path, sr=None)

    target_sr = max(sr_vocal, sr_music)
    if sr_vocal != target_sr:
        vocal = librosa.resample(vocal, orig_sr=sr_vocal, target_sr=target_sr)
    if sr_music != target_sr:
        music = librosa.resample(music, orig_sr=sr_music, target_sr=target_sr)

    if len(music) < len(vocal):
        music = np.pad(music, (0, len(vocal) - len(music)), mode='constant')
    else:
        music = music[:len(vocal)]

    vocal_loudness = measure_loudness(vocal)
    music_loudness = measure_loudness(music)
    target_vocal_db = TARGET_LOUDNESS[mood]["vocals"]
    target_music_db = TARGET_LOUDNESS[mood]["music"]

    gain_vocal = target_vocal_db - vocal_loudness
    gain_music = target_music_db - music_loudness

    print(f"🎛️ Vocal loudness: {vocal_loudness:.2f} dB -> Target: {target_vocal_db:.2f} dB")
    print(f"🎛️ Music loudness: {music_loudness:.2f} dB -> Target: {target_music_db:.2f} dB")

    vocal = apply_gain(vocal, gain_vocal)
    music = apply_gain(music, gain_music)

    mixed = music * 0.6 + vocal * 0.8
    mixed = mixed / (np.max(np.abs(mixed)) + 1e-6) * 0.95

    sf.write(output_path, mixed, target_sr)
    print(f"✅ Final song with melody saved as: {output_path}")
    return output_path

def main():
    lyrics = """
♪ Twinkle twinkle ♫ little star ♪
♫ How I wonder ♫ what you are ♫
"""
    mood = "angry"

    speed_factors = {
        "happy": 1.15, "sad": 0.5, "angry": 1.9,
        "romantic": 0.95, "excited": 1.4, "haunting": 0.85
    }

    pitch_factors = {
        "happy": 0.0, "sad": -2.0, "angry": 3.0,
        "romantic": -1.0, "excited": 1.6, "haunting": -6.0
    }

    speed_factor = speed_factors.get(mood, 1.0)
    pitch_factor = pitch_factors.get(mood, 0.0)

    try:
        final_vocals = generate_vocals_with_melody(lyrics, mood, speed_factor=speed_factor, pitch_factor=pitch_factor)
        if final_vocals:
            bgm_file = generate_music_with_melody(mood, lyrics, speed_factor=speed_factor)
            final_song = mix_vocals_and_music(final_vocals, bgm_file, mood=mood)
            print(f"\n🎵 Final complete song: {final_song}")
    except Exception as e:
        print(str(e))

if __name__ == "__main__":
    main()
```

```python
import librosa
import numpy as np
import soundfile as sf
import parselmouth
import torch
from bark import generate_audio as bark_generate_audio
from bark.generation import preload_models
from audiocraft.models import MusicGen
import difflib
import whisper
import os

TARGET_LOUDNESS = {
    "happy": {"vocals": -12.0, "music": -18.0},
    "sad": {"vocals": -18.0, "music": -20.0},
    "angry": {"vocals": -2.0, "music": -14.0},
    "romantic": {"vocals": -15.0, "music": -18.0},
    "excited": {"vocals": -10.0, "music": -13.0},
    "haunting": {"vocals": -17.0, "music": -19.0}
}

def generate_music_with_melody(mood, lyrics, duration=12, speed_factor=1.0):
    print(f"🎼 Generating {mood} background music with melody for the song...")
    model = MusicGen.get_pretrained('facebook/musicgen-small')
    model.set_generation_params(duration=duration)

    mood_descriptions = {
        "happy": "funky groove with brass instruments and energetic bassline",
        "sad": "slow emotional piano melody",
        "angry": "angry tone",
        "romantic": "soft acoustic melody with gentle rhythm and emotional tone",
        "excited": "fast-paced electronic melody with bright synths",
        "haunting": "dark ambient melody with eerie tones and slow pads"
    }
    description = mood_descriptions.get(mood.lower(), "ambient instrumental")
    wav = model.generate([description])[0].cpu().numpy()

    if not np.all(np.isfinite(wav)):
        print("⚠️ Detected NaNs or infs in generated music. Replacing with zeros.")
        wav = np.nan_to_num(wav)

    if np.max(np.abs(wav)) > 0:
        wav = wav / np.max(np.abs(wav)) * 0.95

    if len(wav.shape) > 1:
        wav = wav[0]

    if speed_factor != 1.0:
        print(f"⏩ Changing speed of background music by factor: {speed_factor}")
        wav = librosa.effects.time_stretch(wav, rate=speed_factor)

    bgm_file = f"{mood}_background_music_with_melody.wav"
    sf.write(bgm_file, wav.astype(np.float32), 32000, format='WAV', subtype='PCM_32')
    print(f"🎧 Background music with melody saved as '{bgm_file}'")
    return bgm_file

def generate_audio_with_bark(lyrics, speed_factor=1.0, pitch_factor=1.0):
    print("🎤 Generating audio using Bark...")
    audio_array = bark_generate_audio(lyrics)

    if speed_factor != 1.0:
        print(f"⏩ Applying speed factor to Bark audio: {speed_factor}")
        audio_array = librosa.effects.time_stretch(audio_array, rate=speed_factor)
    if pitch_factor != 0.0:
        print(f"🎚️ Applying pitch shift to Bark audio: {pitch_factor} semitones")
        audio_array = librosa.effects.pitch_shift(audio_array, sr=24000, n_steps=pitch_factor)

    return audio_array

def generate_vocals_with_melody(lyrics, mood="happy", max_attempts=2, speed_factor=1.0, pitch_factor=1.0):
    preload_models()
    best_score = -1
    best_file = None

    for attempt in range(1, max_attempts + 1):
        print(f"\n🎶 Attempt {attempt} for mood '{mood}'...")
        try:
            audio_array = generate_audio_with_bark(lyrics, speed_factor=speed_factor, pitch_factor=pitch_factor)
        except Exception as e:
            print(f"❌ Bark error: {e}")
            continue

        vocals_file = f"{mood}_vocals_attempt{attempt}.wav"
        sf.write(vocals_file, audio_array, 24000)

        autotuned_file = f"{mood}_vocals_attempt{attempt}_autotuned.wav"
        apply_auto_tune(vocals_file, autotuned_file, mood=mood)

        y, sr = librosa.load(autotuned_file, sr=24000)
        rms = check_noise_level(y)
        silence = check_drag(y, sr)
        dragged = detect_dragged_vowels(y, sr)
        transcription = transcribe_audio(autotuned_file)
        similarity = check_similarity(lyrics, transcription)

        score = (similarity * 2) + (1 - min(silence / 2.0, 1)) + (1 - min(rms / 0.2, 1)) + (1 - min(dragged / 2.0, 1))
        print(f"📊 Final Score: {score:.2f}")

        if score > best_score:
            best_score = score
            best_file = autotuned_file

        if similarity >= 0.75 and silence < 0.6 and rms < 0.1 and dragged < 1.0:
            print("✅ All checks passed!")
            return autotuned_file

    print("\n⚠️ Returning best available attempt.")
    return best_file

def apply_auto_tune(input_file, output_file, mood="happy"):
    snd = parselmouth.Sound(input_file)
    pitch = snd.to_pitch()
    pitch_values = pitch.selected_array["frequency"]
    mean_pitch = np.nanmean(pitch_values[pitch_values > 0])

    target_pitch = {
        "angry": 300,
        "happy": 200,
        "excited": 220,
        "romantic": 190,
        "haunting": 160,
        "sad": 170
    }.get(mood, 180)

    print(f"🎯 Target pitch: {target_pitch} Hz | Measured pitch: {mean_pitch:.2f} Hz")

    if mean_pitch > 0:
        pitch_shift = target_pitch / mean_pitch
        print(f"🔧 Pitch shifting factor applied: {pitch_shift:.2f}")
        snd = snd.resample(snd.sampling_frequency * pitch_shift)

    samples = snd.values[0]
    if np.max(np.abs(samples)) > 0:
        samples = samples / np.max(np.abs(samples)) * 0.9
        snd.values[0] = samples

    snd.save(output_file, "WAV")
    print(f"🎵 Auto-Tuned vocals saved as '{output_file}'")
    return output_file

def check_noise_level(audio_array):
    audio_array = np.array(audio_array)
    audio_array = audio_array / np.max(np.abs(audio_array))
    rms = np.sqrt(np.mean(audio_array**2))
    print(f"🔍 RMS (noise level): {rms:.4f}")
    return rms

def check_drag(audio_array, sr, silence_threshold=0.005, frame_duration=0.01):
    frame_length = int(frame_duration * sr)
    hop_length = frame_length
    envelope = np.array([
        np.max(np.abs(audio_array[i:i+frame_length]))
        for i in range(0, len(audio_array), hop_length)
    ])
    silent_frames = envelope < silence_threshold
    max_silence = 0.0
    current = 0.0
    for is_silent in silent_frames:
        if is_silent:
            current += frame_duration
            max_silence = max(max_silence, current)
        else:
            current = 0.0
    print(f"🧘 Longest silence: {max_silence:.2f} sec")
    return max_silence

def detect_dragged_vowels(y, sr, max_syllable_duration=0.6):
    S, _ = librosa.magphase(librosa.stft(y))
    rms = librosa.feature.rms(S=S)[0]
    times = librosa.times_like(rms, sr=sr)

    sustained = rms > np.percentile(rms, 75)
    sustained_regions = []
    start = None

    for i in range(len(sustained)):
        if sustained[i] and start is None:
            start = times[i]
        elif not sustained[i] and start is not None:
            duration = times[i] - start
            if duration > max_syllable_duration:
                sustained_regions.append(duration)
            start = None

    dragged_duration = sum(sustained_regions)
    print(f"🎤 Detected dragged vowels total duration: {dragged_duration:.2f}s")
    return dragged_duration

def check_similarity(original_text, transcribed_text):
    ratio = difflib.SequenceMatcher(None, original_text.lower(), transcribed_text.lower()).ratio()
    print(f"📝 Similarity ratio: {ratio:.2%}")
    return ratio

def transcribe_audio(file):
    model = whisper.load_model("medium")
    result = model.transcribe(file, language="en")
    print(f"📄 Transcription: {result['text']}")
    return result["text"]

def measure_loudness(audio, ref=1.0):
    rms = np.sqrt(np.mean(audio ** 2))
    db = 20 * np.log10(rms / ref + 1e-9)
    return db

def apply_gain(audio, gain_db):
    factor = 10 ** (gain_db / 20)
    print(f"🎚️ Applying gain: {gain_db:.2f} dB (factor: {factor:.2f})")
    return audio * factor

def mix_vocals_and_music(vocal_path, music_path, output_path="final_song_with_melody.wav", mood="happy"):
    print(f"🎚️ Mixing vocals and background music with melody...")
    vocal, sr_vocal = librosa.load(vocal_path, sr=None)
    music, sr_music = librosa.load(music_path, sr=None)

    target_sr = max(sr_vocal, sr_music)
    if sr_vocal != target_sr:
        vocal = librosa.resample(vocal, orig_sr=sr_vocal, target_sr=target_sr)
    if sr_music != target_sr:
        music = librosa.resample(music, orig_sr=sr_music, target_sr=target_sr)

    if len(music) < len(vocal):
        music = np.pad(music, (0, len(vocal) - len(music)), mode='constant')
    else:
        music = music[:len(vocal)]

    vocal_loudness = measure_loudness(vocal)
    music_loudness = measure_loudness(music)
    target_vocal_db = TARGET_LOUDNESS[mood]["vocals"]
    target_music_db = TARGET_LOUDNESS[mood]["music"]

    gain_vocal = target_vocal_db - vocal_loudness
    gain_music = target_music_db - music_loudness

    print(f"🎛️ Vocal loudness: {vocal_loudness:.2f} dB -> Target: {target_vocal_db:.2f} dB")
    print(f"🎛️ Music loudness: {music_loudness:.2f} dB -> Target: {target_music_db:.2f} dB")

    vocal = apply_gain(vocal, gain_vocal)
    music = apply_gain(music, gain_music)

    mixed = music * 0.6 + vocal * 0.8
    mixed = mixed / (np.max(np.abs(mixed)) + 1e-6) * 0.95

    sf.write(output_path, mixed, target_sr)
    print(f"✅ Final song with melody saved as: {output_path}")
    return output_path

def main():
    lyrics = """
♪ Twinkle twinkle ♫ little star ♪
♫ How I wonder ♫ what you are ♫
"""
    mood = "angry"

    speed_factors = {
        "happy": 1.15, "sad": 0.5, "angry": 1.9,
        "romantic": 0.95, "excited": 1.4, "haunting": 0.85
    }

    pitch_factors = {
        "happy": 0.0, "sad": -2.0, "angry": 3.0,
        "romantic": -1.0, "excited": 1.6, "haunting": -6.0
    }

    speed_factor = speed_factors.get(mood, 1.0)
    pitch_factor = pitch_factors.get(mood, 0.0)

    try:
        final_vocals = generate_vocals_with_melody(lyrics, mood, speed_factor=speed_factor, pitch_factor=pitch_factor)
        if final_vocals:
            bgm_file = generate_music_with_melody(mood, lyrics, speed_factor=speed_factor)
            final_song = mix_vocals_and_music(final_vocals, bgm_file, mood=mood)
            print(f"\n🎵 Final complete song: {final_song}")
    except Exception as e:
        print(str(e))

if __name__ == "__main__":
    main()
```

```python
import numpy as np
import soundfile as sf
import parselmouth
import torch
from bark import generate_audio as bark_generate_audio
from bark.generation import preload_models
from audiocraft.models import MusicGen
import difflib
import whisper_timestamped as whisper
import os
import librosa

TARGET_LOUDNESS = {
    "happy": {"vocals": -12.0, "music": -18.0},
    "sad": {"vocals": -18.0, "music": -20.0},
    "angry": {"vocals": -2.0, "music": -14.0},
    "romantic": {"vocals": -15.0, "music": -18.0},
    "excited": {"vocals": -10.0, "music": -13.0},
    "haunting": {"vocals": -17.0, "music": -19.0}
}

def stretch_audio_by_word(audio, sr, words, factor):
    y_stretched = np.zeros(int(len(audio) * factor) + sr)
    pointer = 0
    for word in words:
        start = int(word['start'] * sr)
        end = int(word['end'] * sr)
        segment = audio[start:end]
        rate = factor
        if len(segment) < 512:
            continue
        S = librosa.stft(segment)
        stretched = librosa.phase_vocoder(S, rate=rate)
        y = librosa.istft(stretched)
        y_stretched[pointer:pointer+len(y)] += y[:len(y_stretched)-pointer]
        pointer += len(y)
    return y_stretched[:pointer]

def generate_music_with_melody(mood, duration):
    print(f"\U0001F3BC Generating {mood} background music with melody...")
    model = MusicGen.get_pretrained('facebook/musicgen-small')
    model.set_generation_params(duration=duration)

    mood_descriptions = {
        "happy": "funky groove with brass instruments and energetic bassline",
        "sad": "slow emotional piano melody",
        "angry": "angry tone",
        "romantic": "soft acoustic melody with gentle rhythm and emotional tone",
        "excited": "fast-paced electronic melody with bright synths",
        "haunting": "dark ambient melody with eerie tones and slow pads"
    }
    description = mood_descriptions.get(mood.lower(), "ambient instrumental")
    wav = model.generate([description])[0].cpu().numpy()

    if not np.all(np.isfinite(wav)):
        print("⚠️ Detected NaNs. Replacing with zeros.")
        wav = np.nan_to_num(wav)

    if np.max(np.abs(wav)) > 0:
        wav = wav / np.max(np.abs(wav)) * 0.95

    if len(wav.shape) > 1:
        wav = wav[0]

    bgm_file = f"{mood}_background_music_with_melody.wav"
    sf.write(bgm_file, wav.astype(np.float32), 32000)
    print(f"🎧 Music saved as '{bgm_file}'")
    return bgm_file

def generate_audio_with_bark(lyrics):
    print("\U0001F3A4 Generating audio using Bark...")
    return bark_generate_audio(lyrics)

def generate_vocals_with_melody(lyrics, mood="happy", speed_factor=1.0, pitch_factor=0.0, attempt=1):
    preload_models()
    audio = generate_audio_with_bark(lyrics)
    vocal_path = f"{mood}_vocals_attempt{attempt}.wav"
    sf.write(vocal_path, audio, 24000)

    print("⏱️ Transcribing for alignment...")
    result = whisper.transcribe(whisper.load_model("base"), vocal_path)
    words = result["segments"][0]['words'] if 'segments' in result and result['segments'] else []

    print("⏩ Stretching vocals word-wise...")
    if abs(speed_factor - 1.0) > 0.05 and words:
        y, sr = librosa.load(vocal_path, sr=24000)
        y = stretch_audio_by_word(y, sr, words, speed_factor)
    else:
        y = audio
        sr = 24000

    if pitch_factor != 0.0:
        print(f"🎚️ Pitch shifting: {pitch_factor} semitones")
        y = librosa.effects.pitch_shift(y, sr=sr, n_steps=pitch_factor)

    final_vocals_file = f"{mood}_final_vocals_attempt{attempt}.wav"
    sf.write(final_vocals_file, y, sr)
    print(f"✅ Vocals saved as '{final_vocals_file}'")
    return final_vocals_file, result['text']

def measure_loudness(audio):
    rms = np.sqrt(np.mean(audio ** 2))
    return 20 * np.log10(rms + 1e-9)

def apply_gain(audio, gain_db):
    return audio * (10 ** (gain_db / 20))

def mix_vocals_and_music(vocal_path, music_path, output_path="final_song.wav", mood="happy"):
    print("\U0001F399️ Mixing...")
    vocal, sr1 = librosa.load(vocal_path, sr=None)
    music, sr2 = librosa.load(music_path, sr=None)
    sr = max(sr1, sr2)
    if sr1 != sr:
        vocal = librosa.resample(vocal, orig_sr=sr1, target_sr=sr)
    if sr2 != sr:
        music = librosa.resample(music, orig_sr=sr2, target_sr=sr)

    if len(music) < len(vocal):
        music = np.pad(music, (0, len(vocal)-len(music)))
    else:
        music = music[:len(vocal)]

    gain_v = TARGET_LOUDNESS[mood]['vocals'] - measure_loudness(vocal)
    gain_m = TARGET_LOUDNESS[mood]['music'] - measure_loudness(music)

    vocal = apply_gain(vocal, gain_v)
    music = apply_gain(music, gain_m)

    mixed = 0.8 * vocal + 0.6 * music
    mixed = mixed / (np.max(np.abs(mixed)) + 1e-6) * 0.95
    sf.write(output_path, mixed, sr)
    print(f"✅ Final song saved: {output_path}")
    return output_path

def get_similarity_score(reference, hypothesis):
    return difflib.SequenceMatcher(None, reference.lower(), hypothesis.lower()).ratio()

def compute_clipping_penalty(audio):
    max_val = np.max(np.abs(audio))
    return 0.0 if max_val <= 1.0 else min(1.0, (max_val - 1.0) * 5)

def compute_loudness_penalty(vocal, music, mood):
    vocal_loud = measure_loudness(vocal)
    music_loud = measure_loudness(music)
    target_v = TARGET_LOUDNESS[mood]['vocals']
    target_m = TARGET_LOUDNESS[mood]['music']
    vocal_dev = abs(vocal_loud - target_v)
    music_dev = abs(music_loud - target_m)
    penalty = (vocal_dev + music_dev) / 20.0
    return min(1.0, penalty)

def compute_smoothness_penalty(audio, sr):
    zcr = librosa.feature.zero_crossing_rate(audio)[0]
    avg_zcr = np.mean(zcr)
    return min(1.0, avg_zcr * 5)

def compute_duration_penalty(audio, sr, target_sec=12.0):
    duration = len(audio) / sr
    dev = abs(duration - target_sec)
    return min(1.0, dev / 5.0)

def compute_total_score(similarity, clipping, loudness, smoothness, duration):
    weights = {
        'similarity': 0.4,
        'clipping': 0.15,
        'loudness': 0.25,
        'smoothness': 0.10,
        'duration': 0.10
    }
    total = (
        weights['similarity'] * similarity +
        weights['clipping'] * (1 - clipping) +
        weights['loudness'] * (1 - loudness) +
        weights['smoothness'] * (1 - smoothness) +
        weights['duration'] * (1 - duration)
    )
    return total

def main():
    lyrics = """
♪ Twinkle twinkle ♫ little star ♪
♫ How I wonder ♫ what you are ♫
"""
    mood = "sad"
    speed = {
        "happy": 1.1,
        "sad": 0.8,
        "angry": 1.7,
        "romantic": 0.85,
        "excited": 1.4,
        "haunting": 0.6
    }.get(mood, 1.0)

    pitch = {
        "happy": 0,
        "sad": -2,
        "angry": 3,
        "romantic": -1,
        "excited": 2,
        "haunting": -3
    }.get(mood, 0.0)

    best_score = -1
    best_output_path = ""
    best_attempt = 0

    for attempt in range(1, 4):
        vocals_path, transcription = generate_vocals_with_melody(
            lyrics, mood=mood, speed_factor=speed, pitch_factor=pitch, attempt=attempt
        )
        # Determine vocal duration
        vocals_audio, sr_voc = librosa.load(vocals_path, sr=None)
        vocal_duration = librosa.get_duration(y=vocals_audio, sr=sr_voc)
        music_path = generate_music_with_melody(mood, duration=vocal_duration)

        similarity = get_similarity_score(lyrics, transcription)
        output_path = f"final_song_attempt{attempt}.wav"
        mix_vocals_and_music(vocals_path, music_path, output_path=output_path, mood=mood)

        final_audio, sr = librosa.load(output_path, sr=None)

        clipping = compute_clipping_penalty(final_audio)
        loudness = compute_loudness_penalty(vocals_audio, vocals_audio, mood)
        smoothness = compute_smoothness_penalty(vocals_audio, sr_voc)
        duration = compute_duration_penalty(final_audio, sr, target_sec=vocal_duration)

        score = compute_total_score(similarity, clipping, loudness, smoothness, duration)

        print(f"\n🎯 Attempt {attempt} Scores:")
        print(f"   - Similarity: {similarity:.4f}")
        print(f"   - Clipping:   {clipping:.4f}")
        print(f"   - Loudness:   {loudness:.4f}")
        print(f"   - Smoothness: {smoothness:.4f}")
        print(f"   - Duration:   {duration:.4f}")
        print(f"   ✅ Total Score: {score:.4f}")

        if score > best_score:
            best_score = score
            best_output_path = output_path
            best_attempt = attempt

    print(f"\n🏆 Best attempt: {best_attempt} with score {best_score:.4f}")
    print(f"🎵 Final selected song: {best_output_path}")

if __name__ == "__main__":
    main()
```
